```python
import matplotlib.pyplot as plt
import numpy as np
from random import randint as rnd
```

```python
plt.xlabel("Yıllar")
plt.ylabel("İhracat")
plt.title("Yıllara göre ihracat rakamlaro")

xpoint = np.array([i for i  in range(2010,2016)])
ypoint = np.array([rnd(800,1200) for i in range(6)])

plt.plot(xpoint,ypoint, "--",marker="D", ms= 30, mec = "y", mfc = "r") #Draw #marker: point type, ms: marker size, mec= dış renk, mfc: tüm renk 
plt.show()
```

```python
font1 = {"family":"aria1","color":"red","size":18}
font2 = {"family":"serif","color":"darkblue","size":15}

plt.title("Accuracy Graph", loc="left")
plt.xlabel("Epoch", fontdict=font1, loc="center")
plt.ylabel("Loss", fontdict=font2, loc="bottom")
plt.grid() #arka kareli


yTrainPoints = np.array([0.8,0.55,0.6,0.55,0.45,0.3])
yValPoint = np.array([0.8,0.65,0.6,0.5,0.6,0.7])

plt.plot(yTrainPoints, linestyle = ":", color="r", lw="2", marker="x")
plt.plot(yValPoint, linestyle = "-", color="y", lw="4", marker="o")
plt.legend(["Train Loss","Value Loss"], loc=4)
plt.show()
```

```python
# x'ler class, y'ler feature

plt.subplot(1,3,1)
plt.plot(yTrainPoints)
plt.title("Train Graph")
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.subplot(1,3,2)
plt.plot(yValPoint)
plt.title("Value Graph")
plt.xlabel("Epoch")
plt.ylabel("Loss")

yTestPoints = np.array([0.65,0.37,0.28,0.33,0.46,0.7])
plt.subplot(1,3,3)
plt.plot(yTestPoints)
plt.title("Test Graph")
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.subplots_adjust(wspace=0.6)  # Grafikler arası yatay boşluğu ayarla
```

```python
x = np.array([rnd(1,100) for i in range(100)])
y = np.array([rnd(1,100) for i in range(100)])

colors = np.array([rnd(1,100) for i in range(100)])
sizes = 10 * np.array([rnd(1,100) for i in range(100)])

plt.scatter(x,y, c = colors, s = sizes, alpha=0.5,
            cmap='nipy_spectral')
plt.colorbar()
plt.show()
```

```python
x = np.array([i for i in range(1,102)])
y = np.array([rnd(1,100) for i in range(1000)])
plt.title("Histogram")

plt.hist(y,x, color="red")
plt.show()
```
